## Bernoulli Distribution - Discrete
       
       
       1. Only two possible outcomes
       
       2. Single Trial

       3. Success= p
          Failure= 1-p=q
         
        
         
         PMF={ q ,x=0   # probability of failure    #x=random variable
               p ,x=1   # probability of success
               
               
          Mean=p,  Variance=pq,  Std.deviation=sq.root(pq) 
          
          
## Normal Distribution(Gaussian Distribution) -Continuous


     Random variable X holds a range of continuous values,  X~ G.D(mean, std)
     
     Std- How far the other elements are distributed from the "mean"
     
     
   Property of Normal Distribution - Follows Empirical Formula
           
             1. 68% of data falls in between the first std.deviation
             
             2. 95% of data falls in between the second std.deviation
             
             3. 99.75% of data falls in between the third std. deviation
             
             
## Binomial Distribution
        
        # Derived from Bernoulli
     
     1. More trials i.e., Multiple Trials   # Difference for Bernoulli and Binomial
     
     2. Each trial is independent
     
     3. Two potential outcomes per trial
     
     4.Probability of Success is same for all trials
